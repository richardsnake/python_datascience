En este notebook vamos a aprender el uso de las librerias para el análisis y visuaizacion de datos

```python
import matplotlib.pyplot as plt
```

## Gráfico de ejemplo

```python
x = [1,2,3,4,5]
y= [5,1,8,9,0]
plt.plot(x,y)
plt.show()
```

## Gráfico de Lineas (Line plot/chart)

```python
anhos = [2018,2019,2020,2021,2022,2023]
temperaturas = [32, 28, 30, 31, 29, 32]
anhos_2 = [2018,2019,2020,2021,2022,2023]
temperaturas_2 = [30, 27, 30, 32, 29, 31]
plt.figure(figsize=(10,6))

plt.plot(anhos,temperaturas, linestyle='-', marker='o', color='g', label="Perú")
plt.plot(anhos_2,temperaturas_2, linestyle='-', marker='o', color='b', label="Chile")

plt.xlabel("Años")
plt.ylabel("Temperatura")
plt.title("Años vs Temperatura")
plt.legend(loc='lower right')

plt.savefig('anhos_temperatura_line.jpg')

plt.show()
```

## Sub gráficos (subplots)

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))

ax[0].plot(anhos,temperaturas, color='g', label="Perú")
ax[0].legend()
ax[1].plot(anhos_2,temperaturas_2, color='b', label="Colombia")
ax[1].legend()
plt.show()
```

## Gráficos de barra

```python
pais = ['Perú', 'Colombia', 'Argentina', 'Ecuador', 'Venezuela', 'Brazil', 'Uruguay', 'Paraguay', 'Bolivia', 'Chile']
copas_americas = [2,1,15, 0, 0, 9,15,2, 1,2]
plt.figure(figsize=(10,6))
plt.bar(pais, copas_americas, color='g')
plt.savefig('pais_vs_copa_america_barras.jpg')
plt.title("Paises y sus copas Américas ganadas")
plt.show()
```

## Gráficos de torta (Pie charts)

```python
#plt.pie(copas_americas, labels=pais, autopct='%1.1f%%')
paises = ['Perú', 'Colombia', 'Argentina']
copas = [3,2,5]
plt.pie(copas, labels=paises)
plt.title("Paises y sus copas Américas ganadas")
plt.legend()
plt.savefig('pais_vs_copa_america_pie.jpg')
plt.show()
```

## Histogramas

```python
edades=[15,16,17,20,21,21,22,23,24,25,26,30,31,32,35]
bins = [15,20,21,30,35]
plt.hist(edades, bins, edgecolor='black')
plt.xlabel('Intervalos de Edades')
plt.ylabel('# de ocurrencias')
plt.title("Histograma de edades")
plt.savefig('Intervalo_edades_histograma.jpg')
plt.show()
```

## Grafico de Caja (Box plot/chart)

```python
plt.boxplot(edades)
plt.show()
```
